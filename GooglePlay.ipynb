Observations
1. Last Updated column we have month, date & year
2. we can segregrate all 3 into columns to get better analysis

```python
df['Updated Month']= df['Last Updated'].str.split(' ').str[0]
df
```

```python
#the last Updated column is divided using split() 
df['Update Date']= df['Last Updated'].str.split(' ').str[1]
df['Update date']= df['Update Date'].str.split(',').str[0]
```

```python
df.drop('Update Date', axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df['Updated Year']= df['Last Updated'].str.split(' ').str[2]
```

```python
#Last Updated column is deleted 
df.drop('Last Updated', axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df.info()
```

```python
#To check if the selected columns has null values
```

```python
df['Updated Month'].isnull().sum()
```

```python
df[df['Update date'].isnull()]
```

```python
df[df['Updated Year'].isnull()]
```

Observations:
1. we got null values in Update date & Updated Year column
2. therefore we were not able to convert obj dtype into int dtype
3. we have to remove or drop the row with null values

```python
#It will erase every row (axis=0) that has "any" Null value in it.
df = df.dropna(how='any',axis=0)
```

```python
df[df['Updated Year'].isnull()]
```

```python
df[df['Update date'].isnull()]
```

```python
df['Update date']= df['Update date'].astype(int)
df['Updated Year']= df['Updated Year'].astype(int) 
```

```python
df.info()
```

Observations:
1. we have dtypes 1 float64, 12 object, 2 int32
2. we need to convert Price & Installs in int32

```python
#to check if column has null values
df['Price'].isnull().sum()
```

```python
df['Installs'].isnull().sum()
```

```python
df
```

```python
df['Price'].str.replace('$', '').str[0]
```

Observations:
1. we were not able to convert Price column to int 32 dtype
2. coz some data has '$' as base so it has to be replaced

```python
df['Price']= df['Price'].str.replace('$', '').str[0]
```

```python
df.head(5)
```

```python
#conversion of Updated Price to int32 dtype
df['Price']= df['Price'].astype(int)
```

```python
df.info()
```

```python
df['Installs']= df['Installs'].str.replace('+', '')
```

Observations:
1. Installs column has '+' & ',' signs by which can not be converted to int32 dtype
2. which has to replaced with blank 

```python
df['Installs']= df['Installs'].str.replace(',', '').astype(int)
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['Reviews'].dtype
```

Observations:
Reviews is in object dtype which can be converted to int32 dtype

```python
df['Reviews']= df['Reviews'].astype(int)
```
